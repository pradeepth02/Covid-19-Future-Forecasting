```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
ss=pd.read_csv("ind-us.csv")
```

```python
ss
```

```python
ss['Date_reported'] = pd.to_datetime(ss['Date_reported'])
```

```python
ss['Year'] = ss['Date_reported'].dt.year
ss['Month'] =ss['Date_reported'].dt.month
ss['Day'] = ss['Date_reported'].dt.day
ss['dayOfWeek'] = ss['Date_reported'].dt.day_name()
ss['WeekOfYear'] = ss['Date_reported'].dt.weekofyear

ss.head(419)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ss['Country']=le.fit_transform(ss['Country'].astype("str"))
ss['Country'].value_counts()
```

```python
y=ss['New_cases']
```

```python
X = ss[['Day','Month','Year','Country']]
```

```python
poly2 = PolynomialFeatures(degree =2)
X_poly = poly2.fit_transform(X)
 
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=15, random_state=0)
regressor.fit(X_poly,y)
```

```python
print(regressor.score(X_poly, y))
```

```python
Day = 23
Month =2
Year =2021
Country =0
#Samples = Samples_31_July[0,0]
X1 = np.array([[Day,Month,Year,Country]])
Total_infections  = regressor.predict(poly2.fit_transform(X1))
```

```python
print('Total No of cases -> {}'.format(int(Total_infections[0])))
```
